```python
#Electricity Consumption Analyzer
print("Goal:   Track Daily Electricity Meter Readings and Compute Daily usage")
print("======================================================================")
print("Problem Statement:")
print("-"*70)
print("You Have Daily Meter Readings, Find:")
print("\t\t1.Daily Consumption")
print("\t\t2.Day with highest usage")
print("\t\t3.Total usage")
print("\t\t4.Plot consumption over time")
print("-"*70)
import pandas as pd
import matplotlib.pyplot as plt
#Meter readings
meter_readings=pd.Series([1020,1035,1050,1070,1085,1100,1115],index=pd.date_range("2025-11-10",periods=7))
print("Meter Readings:\n",meter_readings)
#Daily consumption difference
daily_consumption=meter_readings.diff().fillna(0)
print("\nDaily Consumption:\n",daily_consumption)
#Day with highest usages
print("\nDay with highest usage:",daily_consumption.max(),"on",daily_consumption.idxmax().date())
#Total usages
print("\nTotal Usages:",daily_consumption.sum(),"units")
#Plot consumption over time
meter_readings.plot(kind='bar',title="Daily Electricity Usage",color='orange')
plt.ylabel("Unit Consumed")
plt.show()
```
